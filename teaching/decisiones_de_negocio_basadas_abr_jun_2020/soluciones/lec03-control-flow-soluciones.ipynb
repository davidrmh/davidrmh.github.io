````python
# 1
# (True or False) and False debe ser False ya que
# (True or False) es igual a True
# y así True and False es False
print( (True or False) and False )

# not False != (False or True) es False ya que
# not False es igual a True
# (False or True) es True
# por lo tanto tenemos True != True lo que es False
print (not False != (False or True))

# not not False and not False es False ya que
# not not False es igual a not True que es igual a False
# not False es True
# por lo tanto tenemos False and True que es False
print(not not False and not False)
````

````python
# 1
temp = float(input('Dame la temperatura'))

if temp >= 35:
    print('Hace mucho calor')
elif temp < -15:
    print('Hace mucho frío')
elif  25 < temp < 35:
    print('Hace calor')
elif -15 <= temp < 12:
    print('Hace frío')
elif 12 <= temp <= 25:
    print('El clima es templado')
````

````python
# 2 
user_pass = {'usr503': '2vu2bo',
'usr085': 'geeaa',
'usr406': 'xqzbiy',
'usr182': 'jbngo0',
'usr168': 'qih6e',
'usr900': '6ynym',
'usr542': '7p6mnd',
'usr847': 'ruqq6y',
'usr629': '9qs9g5',
'usr418': 'f15lg'}

usr = input('Dame un usuario')

if usr in user_pass:
    print('Usuario registrado')
else:
    print('Usuario no registrado')
````

````python
# 3
user_pass = {'usr503': '2vu2bo',
'usr085': 'geeaa',
'usr406': 'xqzbiy',
'usr182': 'jbngo0',
'usr168': 'qih6e',
'usr900': '6ynym',
'usr542': '7p6mnd',
'usr847': 'ruqq6y',
'usr629': '9qs9g5',
'usr418': 'f15lg'}

usr = input('Dame un usuario')

if usr in user_pass:
    #usuario registrado
    print('Usuario registrado')
    
else:
    #Usuario no registrado
    print('Usuario no registrado')
    resp = input('¿Deseas registrar al usuario?')
    
    if resp.lower() == 's':
        pwd = input('Dame una contraseña')
        #Agrega al diccionario
        user_pass[usr] = pwd
    elif resp.lower() == 'n':
        print('El usuario no fue registrado')
    else:
        print('Opción no válida. Saliendo del programa')    
        
print('-' * 50)
print(user_pass)
````
