Cree un diccionario que contenga el conteo de cada palabra del siguiente texto
```python
texto = '''
Tres tristes tigres tragaban trigo en un trigal en un trigal tragaban trigo
tres tristes tigres estos Tigres tragaban mucho trigo en un trigal
'''
```

````python
texto = '''
Tres tristes tigres tragaban trigo en un trigal en un trigal tragaban trigo
tres tristes tigres estos Tigres tragaban mucho trigo en un trigal
'''

# Primero convertimos a minúsculas
texto = texto.lower()

#hacemos un split
lista_texto = texto.split()

#Las llaves del diccionario serán palabras (strings)
#los valores el número de veces que aparece es palabra
dic_conteo = dict()

for palabra in lista_texto:
    
    #Si la palabra no estaba en el diccionario
    #Se hace un primer conteo
    if palabra not in dic_conteo:
        dic_conteo[palabra] = 1
        
    #Si ya estaba, entonces se incrementa el conteo    
    else:
        dic_conteo[palabra] = dic_conteo[palabra] + 1 

print(dic_conteo)
````

````python
# Otra forma
dic_conteo = dict()

for palabra in lista_texto:
    #El método count de las listas
    #cuenta el número de veces que ocurre
    #el argumento enviado (en este caso la palabra)
    print(palabra)
    dic_conteo[palabra] = lista_texto.count(palabra)

print(dic_conteo)    
````

````python
# 1
lista_tasas = ['0.05', '0.1', '0.15', '0.20']
lista_cupones = [5, 10, 15, 10]
nominal = 10

#obtiene el número de cupones
n = len(lista_tasas)

#para guardar el precio
precio = 0

for t in range(n):
    #Las tasas son string
    #es necesario hacer el casting
    tasa = float(lista_tasas[t])
    
    cupon = lista_cupones[t]
    
    #agrega el valor presente del cupón
    precio = precio + cupon / (1 + tasa)**(t + 1)

#se agrega el valor presente del nominal
precio = precio + nominal / (1 + tasa)**(n)

print('El precio es', round(precio, 2))
````
